```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

# Başlangıç ve bitiş tarihlerini tanımlayın
baslangic_tarihi = datetime(2013, 9, 1)
bitis_tarihi = datetime(2023, 9, 1)

# Boş bir liste oluşturun
url_list = []

# Tarih aralığındaki her gün için link oluşturun
while baslangic_tarihi <= bitis_tarihi:
    tarih_str = baslangic_tarihi.strftime('%d-%B-%Y')
    link = f"https://tr.weatherspark.com/h/d/95434/{baslangic_tarihi.strftime('%Y/%m/%d')}/{tarih_str}-tarihinde-in%C4%B0stanbul-T%C3%BCrkiye-Ortalama-Hava-Durumu#Figures-Temperature"
    url_list.append(link)
    baslangic_tarihi += timedelta(days=1)
```
