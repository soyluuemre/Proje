```python
import pandas as pd

# Dosya yollarını içeren liste
file_paths = ['2019data.csv', '2020data.csv', '2021data.csv', '2022data.csv', '2023data.csv']

# Boş bir DataFrame oluşturun
combined_df = pd.DataFrame()

# Her bir CSV dosyasını okuyun ve birleştirin
for file_path in file_paths:
    df = pd.read_csv(file_path)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Sonuçları görüntüleyin
print(combined_df)
```

```python
combined_df
```

```python
# Belirtilen sütunlarda NaN değerleri kontrol ederek, hepsi NaN olan satırları kaldırın
cols_to_check = ['Dew_Point', 'TD1', 'TD2', 'TD3', 'TD4', 'TD5']
combined_df = combined_df.dropna(subset=cols_to_check, how='all')
```

```python
# Yeni bir DataFrame oluşturun, bu, NaN değerleri kaldırılmış verileri içerecektir
df = combined_df.dropna(subset=cols_to_check, how='all').copy()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

df
```

```python
# "TD3" sütunundaki her değeri işleyin ve sadece sayısal karakterleri alın
def extract_numerical_value(text):
    numerical_value = ''.join(filter(lambda x: x.isdigit() or x in [',', '.'], text))
    return numerical_value

df['TD3'] = df['TD3'].apply(lambda x: extract_numerical_value(str(x)))

df
```

```python
import re

# TD5 sütunundaki her bir değeri düzenleyin
def clean_td5(td5_value):
    # Parantez içindeki ifadeleri kaldırın
    cleaned_value = re.sub(r'\([^)]*\)', '', td5_value)
    # İki ardışık virgülü tek bir virgüle dönüştürün ve baştaki ve sondaki boşlukları kaldırın
    cleaned_value = re.sub(r',+', ',', cleaned_value).strip()
    return cleaned_value

# TD5 sütunundaki her bir değeri düzenleyin ve sonucu yeni bir sütuna ekleyin
df['Düzenlenmiş_TD5'] = df['TD5'].apply(clean_td5)

# Sonucu görüntüleyin
print(df)
```

```python
# "TD5" sütunundaki her bir değeri düzenleyin
def clean_td5(Düzenlenmiş_TD5_value):
    # Parantez içindeki ifadeleri kaldırın
    cleaned_value = re.sub(r'\([^)]*\)', '', Düzenlenmiş_TD5_value)
    # İki ardışık boşluğu tek bir boşluğa dönüştürün ve baştaki ve sondaki boşlukları kaldırın
    cleaned_value = re.sub(r'\s+', ' ', cleaned_value).strip()
    return cleaned_value

# "TD5" sütununu düzenleyin
df['Düzenlenmiş_TD5'] = df['Düzenlenmiş_TD5'].apply(clean_td5)

# Yeni bir sütun oluşturun ve her bir değeri sayısal bir değere dönüştürün
mapping = {
    "Bulut Tavanı ve Görüş Temiz": 1,
    "Çok bulutlu": 2,
    "Parçalı Bulutlu": 3,
    "Az Bulutlu": 4,
    "Önemli Bulut Yok": 5,
    "Kapalı": 6
}
df['Bulut Durumu'] = df['Düzenlenmiş_TD5'].map(mapping)

# Sonucu görüntüleyin
print(df[['Düzenlenmiş_TD5', 'Bulut Durumu']])
```

```python
df
```

```python
# Sütun adlarını değiştirin
df = df.rename(columns={
    'TD1': 'Derece',
    'TD2': 'Yükseklik Ölçer',
    'TD3': 'Rüzgar Hızı',
    'TD4': 'Görünürlük',
    'Düzenlenmiş_TD5': 'Bulut'
})

# Sonucu görüntüleyin
print(df)
```

```python
df
```

```python
# Dönüşüm için bir işlev tanımlayın
def dönüştür_bulut(değer):
    if değer == 'Bulut Tavanı ve Görüş Temiz':
        return '1'
    elif değer == 'Çok bulutlu':
        return '2'
    elif değer == 'Parçalı Bulutlu':
        return '3'
    elif değer == 'Az Bulutlu':
        return '4'
    elif değer == 'Önemli Bulut Yok':
        return '5'
    elif değer == 'Kapalı':
        return '6'
    else:
        return değer  # Diğer durumlar için aynı değeri koru

# 'Bulut' sütununu kullanarak yeni bir sütun oluşturun
df['Yeni_Bulut'] = df['Bulut'].apply(dönüştür_bulut).astype(str)

# Sonuçları görüntüleyin
print(df)
```
